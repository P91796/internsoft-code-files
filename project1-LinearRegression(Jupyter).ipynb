```python
#importing the libraries
import pandas as pd
import matplotlib.pyplot as plt

#this allows plots to appear directly in notebook 
%matplotlib inline
```

```python
#read into DataFrame
data=pd.read_csv('advertising.csv')
data.head()
```

```python
#shape of DataFrame
data.shape
```

```python
#visualising the relationship between features and response using scatterplots
fig,axs= plt.subplots(1,3,sharey=True)
data.plot(kind='scatter',x='TV',y='Sales',ax=axs[0],figsize=(16,8))
data.plot(kind='scatter',x='Radio',y='Sales',ax=axs[1])
data.plot(kind='scatter',x='Newspaper',y='Sales',ax=axs[2])
```

```python
#creating x and y
feature_cols=['TV']
X=data[feature_cols]
y=data.Sales

#sklearn
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X,y)

#print intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
#calculate the prediction
6.97482 + 0.055467*50
```

```python
#creating a DataFrame having
X_new=pd.DataFrame({'TV':[50]})
X_new.head()
```

```python
# use the model to make predictions on a new value
lm.predict(X_new)
```

```python
# create a DataFrame with the min and max values of TV
X_new=pd.DataFrame({'TV': [data.TV.min(), data.TV.max()]})
X_new.head()
```

```python
# make predictions for those x values and store 
preds = lm.predict(X_new)
preds
```

```python
#plot the observed data
data.plot(kind='scatter',x='TV',y='Sales')

#plot the least squares line
plt.plot(X_new,preds,c='red',linewidth=3)
```

```python
import statsmodels.formula.api as smf
lm=smf.ols(formula='Sales ~ TV',data=data).fit()
lm.conf_int()
```

```python
#print p-values for model coeffients
lm.pvalues
```

```python
#print r-squared value for model
lm.rsquared
```

```python
#creating x and y for multi LinearRegression
feature_cols=['TV','Radio','Newspaper']
X=data[feature_cols]
y=data.Sales

lm=LinearRegression()
lm.fit(X,y)

#print intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
lm=smf.ols(formula='Sales ~ TV+Radio+Newspaper',data=data).fit()
lm.conf_int()
lm.summary()
```

```python
lm = smf.ols(formula='Sales ~ TV + Radio', data=data).fit()
lm.rsquared
```

```python
lm = smf.ols(formula='Sales ~ TV + Radio + Newspaper', data=data).fit()
lm.rsquared
```
